<p>When we try to add elements with diffrent datatype in array all elements get automatically converted to an single datatype</p>

```python
arr3 = np.array(["S", 2, 3, 4, 5, {1, 2, 3, 4}])
print(arr3)
```

```python
print(arr3[5])
```
