## Numpy Indexing and Slicing of Vectors

```python
arr = np.arange(11, 21)
arr
```

```python
arr[7]
```

```python
arr[6]
```

```python
arr[1:5]
```

```python
arr[:]
```

```python
arr[2:]
```

```python
arr[:2]
```

```python
arr[2:8:2]
```

### Numpy Indexing & Slicing - MATRIX

```python
arr = np.arange(1, 30+1).reshape(6, 5)
arr
```

```python
arr[5] #Single for rowsS
```

```python
arr[0,0] #accesing first element 1st row & first column
```

```python
arr[5,4]
```

```python
s = arr[0:2, 1:3]
s
```

```python
arr[3:6, 3:5] # or arr[3:, 3:]
```

```python
arr[:,3] #Accessing Column
```

### Boolean Indexing

```python
arr = np.arange(11, 21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
