#### Stacking of Array
- Cannot be done on the matrix
- only done on array 1D arrays

```python
A = np.array([1, 2, 3, 4])
B = np.array([5, 6, 7, 8])
```

```python
np.vstack((A, B))
```

```python
np.hstack((A, B))
```

```python
np.column_stack((A, B))
```

#### Spliting Arrays

```python
c = np.arange(16).reshape(4, 4)
c
```

```python
np.hsplit(c, 2)
```

```python
np.vsplit(c, 4)
```
