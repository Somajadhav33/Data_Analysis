```python
import numpy as np
```

#### Stacking of Array
- Cannot be done on the matrix
- only done on array 1D arrays

```python
A = np.array([1, 2, 3, 4])
B = np.array([5, 6, 7, 8])
```

```python
np.vstack((A, B))
```

```python
np.hstack((A, B))
```

```python
np.column_stack((A, B))
```

#### Spliting Arrays

```python
c = np.arange(16).reshape(4, 4)
c
```

```python
np.hsplit(c, 2)
```

```python
np.vsplit(c, 4)
```

#### Searching in Array

```python
arr = np.array([1, 2, 3, 4, 5, 4, 4])
arr
```

```python
x = np.where(arr == 4)
x
```

```python
arr = [6, 7, 8, 9]
x = np.searchsorted(arr, 7)
y = np.searchsorted(arr, 7, side="right")
print(x)
print(y)
arr = np.array([1, 3, 5, 7])
x = np.searchsorted(arr, [2, 4, 6])
x
```

```python
arr = np.array([1, 2, 3, 4, 5, 4, 4])
np.sort(arr)
```

```python
arr = np.array([[4, 5, 6, 3], [6, 4, 3, 2]])
np.sort(arr)
```
