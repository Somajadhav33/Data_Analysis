### Array Operations

```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])
```

```python
arr1 + arr2
```

```python
arr2 - arr1
```

```python
arr1 * arr2
```

```python
arr1 / arr2
```

```python
arr1 // arr2
```

```python
arr1 ** arr2
```

```python
arr2 = np.arange(1, 26).reshape(5,5)
arr2
```

```python
arr2 + 100
```

### BroadCasting

```python
l = [10,20,30,40]
arr = np.array(l)
```

```python
arr + 10
```

### Deep and Shallow Copy

```python
a = np.arange(1,11)
a
```

```python
sli = a[:5]
sli * 10
```

```python
a[4:6] * 10
```

```python
b = a.copy()
```

```python
b
```

```python
b[0] = 11
```

```python
b
```
